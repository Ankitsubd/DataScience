#### Boolean or Mask Indexing

```python
e = np.random.randint(1,7,(5,5))
e
```

```python
e >= 4 # check it's is greater or equal to 4 if yes return True otherwise False 
```

```python
mask = e >= 4
```

```python
e[mask] #return only true value
```

```python
mask = [True, False, True, False, True]
```

```python
e
```

```python
e[mask]
```
